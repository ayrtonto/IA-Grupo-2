Importante, instalar pyarrow y pandas para poder ver el dataset

**Esta parte solo sirvio para poder borrar algunos datos que no otorgaban informacion valiosa, el data set nuevo se llama dataset.parquet, el original se puede encontrar en la pagina web HugginFace: https://huggingface.co/datasets/somosnlp/spanish_medica_llm**

```python
import pandas as pd
```

```python
df = pd.read_parquet('./raw_dataset.parquet')
```

```python
df.info()
```

```python
df_filtrado = df
```

```python
# Visualizacion segun fuente
df_filtrado[(df_filtrado["source"] == '1')]
```

```python
# Eliminacion de los datos feos
df_filtrado = df_filtrado.drop(df_filtrado[df_filtrado['source'] == '2'].index)
df_filtrado = df_filtrado.drop(df_filtrado[df_filtrado['source'] == '3'].index)
df_filtrado = df_filtrado.drop(df_filtrado[df_filtrado['source'] == '4'].index)
df_filtrado = df_filtrado.drop(df_filtrado[df_filtrado['source'] == '7'].index)
df_filtrado = df_filtrado.drop(df_filtrado[df_filtrado['source'] == '8'].index)
df_filtrado = df_filtrado.drop(df_filtrado[df_filtrado['source'] == '13'].index)
```

```python
# Visualizacion segun documento y fuente
df_filtrado[(df_filtrado["document_id"] == '1') & (df_filtrado["source"] == '2')]
```

```python
df_filtrado.describe()
```

```python
df_filtrado.to_parquet("dataset.parquet")
```
